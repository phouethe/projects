```python
## import libraries
import pandas as pd
```

```python
## reading in peripheral dataset, also selecting an easier variable to type
gear = pd.read_csv("peripherals.csv")
```

```python
## taking only the columns I am interested from this dataset
gear = gear[["Player", "Mouse", "eDPI", "Mousepad", "Keyboard"]]
```

```python
gear.head()
```

```python
## read in player dataset
players = pd.read_csv("valorant_player_stats.csv")
```

```python
players
```

```python
## delete unnecessary column
del players["Unnamed: 0"]
```

```python
players
```

```python
## make the gear player names lowercase
gear["Player"] = gear["Player"].str.lower()

## create new column of lowercase names in players
players["playersLower"] = players["player"].str.lower()
```

```python
players.tail()
```

```python
## merging datasets
## had to do a left_on and right_on merge as the player columns were not technically named the same

combined = pd.merge(players,
                   gear,
                   left_on="playersLower",
                   right_on="Player",
                   how='left')
```

```python
combined.tail()
```

```python
## dropping redundant column after merging

combined = combined.drop('Player', axis=1)
combined = combined.drop('playersLower', axis=1)
```

```python
combined.tail()
```

```python
## save as csv to be used in data viz

combined.to_csv("valorant_players.csv")
```
